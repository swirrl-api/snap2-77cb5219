```python
# First we specify Coordinate Refference System for Map Projection
# We will use Mercator, which is a cylindrical, conformal projection. 
# It has bery large distortion at high latitudes, cannot 
# fully reach the polar regions.
projection = ccrs.Mercator()

# Specify CRS, that will be used to tell the code, where should our data be plotted
crs = ccrs.PlateCarree()

# Now we will create axes object having specific projection 
plt.figure(dpi=150)
ax = plt.axes(projection=projection, frameon=True)

# Draw gridlines in degrees over Mercator map
gl = ax.gridlines(crs=crs, draw_labels=True,
                  linewidth=.6, color='gray', alpha=0.5, linestyle='-.')
gl.xlabel_style = {"size" : 7}
gl.ylabel_style = {"size" : 7}

# To plot borders and coastlines, we can use cartopy feature
ax.add_feature(cf.COASTLINE.with_scale("50m"), lw=0.5)
ax.add_feature(cf.BORDERS.with_scale("50m"), lw=0.3)

dataset = nc.Dataset("../data/bc_2003.nc")

sst = dataset.variables['BC'][56, :, :] #52
lats = dataset.variables['lat'][:]
lons = dataset.variables['lon'][:]

plt.contourf(lons, lats, sst, 1,
             transform=ccrs.PlateCarree())

plt.show()
```

# Distributions 2003

## BC

```python
daily_average_bc = []

for i in range(bc_2003.sizes['time']):
    # Select data for the chosen day
    daily_data = bc_2003.isel(time=i)

    # Calculate the average across all latitudes and longitudes
    average_value = daily_data.mean(dim=("lat", "lon"))

    value = average_value.BC.values
    
    daily_average_bc.append(float(value))
```

```python
plt.plot(daily_average_bc)
```

## CC

```python
daily_average_cc = []

for i in range(cc_2003.sizes['time']):
    # Select data for the chosen day
    daily_data = cc_2003.isel(time=i)

    # Calculate the average across all latitudes and longitudes
    average_value = daily_data.mean(dim=("lat", "lon"))

    value = average_value.CC.values
    
    daily_average_cc.append(float(value))
```

```python
plt.plot(daily_average_cc)
```

```python
days = []
for i in range(cc_2003.sizes['time']):
    days.append(i)

data = {'BC': daily_average_bc, 
       'CC': daily_average_cc,
       'DC': daily_average_dc,
       'ID': daily_average_id,
       'OD': daily_average_od,
       'day': days}

df = pd.DataFrame(data)
```

```python
test = pd.melt(df, id_vars='day', value_vars=['BC', 'CC', 'DC', 'ID', 'OD'])
```

```python
(
    ggplot(test)  # What data to use
    + aes(x="day", y="value", color='variable')  # What variable to use
    + geom_line()  # Geometric object to use for drawing
    + labs(title = "Day related to the mean peak density",
        subtitle = "",
        x = "Years",
        y = "Mean",
        caption = "Linear regression line shown with 95% confidence band") + 
    coord_cartesian(ylim=(0, 0.2))
)
```
