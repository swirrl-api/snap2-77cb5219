# Trend analysis (joy division)

```python
# Load libraries
import pymannkendall as mk
from scipy.stats import linregress
import pandas as pd
from plotnine.data import economics
from plotnine import ggplot, aes, geom_line, geom_point, geom_smooth, labs
from matplotlib import pyplot as plt
import seaborn as sns
from scipy.signal import find_peaks
import numpy as np
```

## Onset

```python
df_joy = pd.read_csv("../data/df_merged_ih.csv")
```

```python
mean = []
year = []
average = []
old = df_joy["1990"].mean()

for i in range(1991, 2021):
    new = df_joy[str(i)].mean()
    mean.append(new)
    year.append(i)
    average.append(((new-old)/old)*100)
```

```python
data = {'Year': year, 
       'Mean': mean, 
       'Average since 1990': average}
```

```python
df_mean = pd.DataFrame(data)
df_mean.to_csv("../data/mean_ih.csv", index=False)
```

```python
df_mean = pd.read_csv("../data/mean_ih.csv")
```

```python
(
    ggplot(df_mean)  # What data to use
    + aes(x="Year", y="Mean")  # What variable to use
    + geom_point()  # Geometric object to use for drawing
    + geom_smooth(method='ols',
                 se = True,
                 level = 0.95) 
    + labs(title = "Day related to the mean peak density",
        subtitle = "",
        x = "Years",
        y = "Mean",
        caption = "Linear regression line shown with 95% confidence band")
)
```

```python
# Fit a linear regression model
slope, intercept, r_value, p_value, std_err = linregress(df_mean['Year'], df_mean['Mean'])

if p_value < 0.05:
    print(f"Statistically significant slope ({slope}).")
else:
    print(f"No significant slope detected ({p_value} > 0.05).")
```

## Highest intensity

```python
def get_peak_location(i, df):
    x = sns.kdeplot(df[str(i)]).lines[0].get_xdata() # Get the x data of the distribution
    y = sns.kdeplot(df[str(i)]).lines[0].get_ydata() # Get the y data of the distribution

    # Find peaks
    i_peaks, _ = find_peaks(y)
    
    # Find the index from the maximum peak
    i_max_peak = i_peaks[np.argmax(y[i_peaks])]
    
    # Find the x value from that index
    x_max = x[i_max_peak]

    plt.close()

    return x_max
```
